网页请求示例：请求新闻首页

````python

import requests

def fetch_news_page():
    url = 'https://news.china.com/domestic/index.html'
    response = requests.get(url)
    response.encoding = 'utf-8' # 解决中文乱码问题
    return response.text

# 调用函数，并打印结果
print(fetch_news_page())
    
````

网页请求示例：请求OpenAI的sora网页

````python
import requests

proxies = {
    'http': 'http://127.0.0.1:7890',
    'https': 'http://127.0.0.1:7890',
}

def fetch_sora_page():
    url = 'https://openai.com/index/sora/'
    response = requests.get(url, proxies=proxies)
    return response.text

# 调用函数，并打印结果
print(fetch_sora_page())
    
````

### 爬虫（获取网页中想要的信息）

这一部分将教你如何解析HTML，提取网页中的信息。

````python
import re
# 调用上面的fetch_news_page函数，获取网页内容
page_text=fetch_news_page() 

# 数据提取
# 使用正则表达式提取新闻时间、标题、来源和链接
# 正则表达式的使用方法可以参考：https://www.runoob.com/python/python-reg-expressions.html
news_time = re.findall(
    r'</em><em class="item_time">(.*)</em></span></li>', page_text)
news_title = re.findall(
    r'target="_blank">(.*)</a></h3><span class="item_info">', page_text)
news_source = re.findall(
    r'<em class="item_source">(.*)</em><em class="item_time">', page_text)
news_url = re.findall(
    r'<h3 class="item_title"><a href="(.*)" target="_blank">', page_text)

print(news_time)
print(news_title)
print(news_source)
print(news_url)
````
